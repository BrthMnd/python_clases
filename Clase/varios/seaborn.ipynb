# SEABORN

## importaciones

```python
import numpy as np
import seaborn as sb
import pandas as pd

import matplotlib.pyplot as plt
```

```python
Excel = pd.read_excel("Estadistica.xlsx",sheet_name="Cuantitativos_Discretos")
Excel.describe()
```

## Creacion de Frecuencia

### value counts with index

```python
frecuencia = Excel["Talla_Zapatos"].value_counts().reset_index()
```

### Frecuencia adsoluta

- Aclarar la columna antes de mostrar datos

```python
frecuencia.columns = ['Talla_Zapatos', 'Frecuencia_Absoluta']
frecuencia
```

### Frecuencia adsoluta acumulada

```python
frecuencia["Frecuencia_Acumulada_Absoluta"] = frecuencia['Frecuencia_Absoluta'].cumsum()
frecuencia
```

### Frecuencia Relativa
- Calcula la frecuencia relativa (frecuencia absoluta / suma de todas las frecuencias absolutas)

```python

suma_absoluta = frecuencia["Frecuencia_Absoluta"].sum()
frecuencia["Frecuencia_Relativa"] = frecuencia["Frecuencia_Absoluta"] / suma_absoluta
frecuencia
```

### frecuencia relativa acumulada

```python
frecuencia['Frecuencia_Acumulada_Relativa'] = frecuencia['Frecuencia_Relativa'].cumsum()
```

# Mostrar 

```python
frecuencia
```

## Grafico de Talla de zapatos con frecuencia adsoluta

```python
plt.figure(figsize=(10,5))  # Ajusta el tamao de la figura si es necesario

# Utiliza Seaborn para crear el grfico de barras
sb.barplot(data=frecuencia, x="Talla_Zapatos", y="Frecuencia_Absoluta")
sb.set_theme(style="whitegrid" )
# Agrega etiquetas y ttulo
plt.xlabel("Talla de zapatos")
plt.ylabel("Frecuencia Adsoluta")
plt.title("Cuantitativos discretos")

# Muestra el grfico
plt.savefig("./img/Cuantitativos discretos - Talla de zapatos.png") 
plt.show()
```

```python
plt.figure(figsize=(15,5))  
sb.set_theme(style="whitegrid" )
# Agrega etiquetas y ttulo

g = sb.displot(data=frecuencia, x="Frecuencia_Absoluta", y="Frecuencia_Acumulada_Absoluta")

# Agrega ttulo
g.fig.suptitle("Grfico lmplot")

# Guarda el grfico lmplot como una imagen (por ejemplo, en formato PNG)

plt.show()
```
