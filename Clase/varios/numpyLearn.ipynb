```python
data = np.random.randint(0,100,(5,5))

# Aplanar la matriz en un array unidimensional
flattened_data = data.flatten()

x_indices = np.arange(flattened_data.shape[0])

print(data)
plt.plot(x_indices, flattened_data, marker='o')
plt.title("Prueba de representacion lineal")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()



```
